```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
target_dir = './download/training_data/'
```

```python
rand = random.randint(0, 49)
df = pd.read_csv(target_dir + 'target' + str(rand) + '.csv')
df
```

```python
def plot(datas):
    plt.figure(figsize=(20, 8))
    plt.plot(range(len(datas)), datas)
    plt.show()
```

```python
plot(df['generation'])
```

```python
plot(df['consumption'])
```

```python
def get_series_mean(series):
    return np.mean(series.to_numpy().reshape((-1, 24)), axis=1)

gen = get_series_mean(df['generation'])
con = get_series_mean(df['consumption'])
```

```python
plot(gen)
```

```python
plot(con)
```

```python
import matplotlib.pyplot as plt
```

```python
keyword = 'generation'
gen_list = []
with open('./training_detail.txt') as f:
    line = f.readline()
    while line:
        if keyword in line:
            a = float(line.split(': ')[1].strip())
            gen_list.append(a)
        line = f.readline()
    f.close()
plt.plot(range(300), gen_list)
plt.xlabel('epoch')
plt.ylabel('loss')
plt.title('Generation Loss')
plt.savefig('Generation_loss.png')
```

```python
keyword = 'consumption'
con_list = []
with open('./training_detail.txt') as f:
    line = f.readline()
    while line:
        if keyword in line:
            a = float(line.split(': ')[1].strip())
            con_list.append(a)
        line = f.readline()
    f.close()
plt.plot(range(300), con_list)
plt.xlabel('epoch')
plt.ylabel('loss')
plt.title('Consumption Loss')
```
